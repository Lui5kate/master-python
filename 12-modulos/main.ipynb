Podemos conseguir modulos que ya vienen en el lenguaje, en internet o crear los propios

```python
# Importar modulos propios
import mimodulo
from mimodulo import holaMundo
from mimodulo import *

print(mimodulo.holaMundo("Luis Lira"))
print(mimodulo.calculadora(3,5,True))

print(holaMundo("Luis Lira"))
```

```python
# Módulo Fechas
import datetime

print(datetime.date.today())

fecha_completa = datetime.datetime.now()
print(fecha_completa)
print(fecha_completa.year)
print(fecha_completa.month)
print(fecha_completa.day)

fecha_personalizada = fecha_completa.strftime("%d/%m/%Y, %H:%M:%S")
print(fecha_personalizada)

print(datetime.datetime.now().timestamp())
print(datetime.datetime.now().time())
```

```python
# Modulo matemáticas
import math

print("La Raíz cuadrada de 9: ", math.sqrt(9))
print("Numero pi: ", float(math.pi))
print("Redondear: ", math.ceil(6.1656453))
print("Redondear: ", math.floor(6.1656453))

# Modulo random
import random

print("Numero aleatorio entre 15 y 67: ", random.randint(15,67))
```
